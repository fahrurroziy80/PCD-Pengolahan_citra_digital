NumPy adalah pustaka Python yang digunakan untuk bekerja dengan array.Ini juga memiliki fungsi untuk bekerja dalam domain aljabar linier, transformasi Fourier, dan matriks.NumPy dibuat pada tahun 2005 oleh Travis Oliphant. Ini adalah proyek sumber terbuka dan Anda dapat menggunakannya secara bebas.NumPy merupakan singkatan dari Numerical Python.

# Why Use NumPy?

Dalam Python, kita memiliki daftar yang berfungsi sebagai array, tetapi mereka lambat dalam pemrosesan.NumPy bertujuan untuk menyediakan objek array yang hingga 50x lebih cepat daripada daftar Python tradisional.Objek array dalam NumPy disebut ndarray, yang menyediakan banyak fungsi pendukung yang membuat bekerja dengan ndarray sangat mudah.Array sangat sering digunakan dalam ilmu data, di mana kecepatan dan sumber daya sangat penting.


```python
import numpy as np
```

```python
x = np.array(list2D, np.int32)
print(x)
```
